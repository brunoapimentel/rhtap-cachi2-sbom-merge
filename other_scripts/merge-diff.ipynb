```python
import json

path = "pip-e2e-test"

with open(f"{path}/cachi2.bom.json") as file:
    cachi2_sbom = json.load(file)

with open(f"{path}/rhtap.bom.json") as file:
    rhtap_sbom = json.load(file)
```

```python
print("Cachi2 reported components")
print(len(cachi2_sbom["components"]))

print("Syft reported components")
print(len(rhtap_sbom["components"]))
```

```python
cachi2_indexed_components = {
    f"{component['name']}@{component.get('version', '')}": component
    for component in cachi2_sbom["components"]
}

rhtap_indexed_components = {
    f"{component['name']}@{component.get('version', '')}": component
    for component in rhtap_sbom["components"]
}
```

```python
syft_unique_components = []

for key in rhtap_indexed_components:
    if key not in cachi2_indexed_components.keys():
        syft_unique_components.append(rhtap_indexed_components[key])
```

```python
import pandas
non_rpms = [component for component in syft_unique_components if "rpm" not in component.get("purl", "")]

df = pandas.DataFrame(non_rpms)

with pandas.option_context('display.max_rows', None):
    display(df[["name", "version", "purl"]])
```
