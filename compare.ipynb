```python
import json
import pandas

with open("cachi2.bom.json") as file:
    cachi2_sbom = json.load(file)

with open("rhtap.bom.json") as file:
    rhtap_sbom = json.load(file)

cachi2_df = pandas.DataFrame(cachi2_sbom["components"])
rhtap_df = pandas.DataFrame(rhtap_sbom["components"])
```

```python
# RHTAP reported modules
df = rhtap_df[rhtap_df["purl"].str.contains("golang", na=False)].reset_index()
df[["name", "version", "purl"]].sort_values(by="name")
```

```python
# RHTAP reported modules - deduped
df = rhtap_df[rhtap_df["purl"].str.contains("golang", na=False)].drop_duplicates(subset="purl").reset_index()
df[["name", "version", "purl"]].sort_values(by="name")
```

```python
# Cachito reported modules
df = cachi2_df[cachi2_df["purl"].str.contains("type=module", na=False)].reset_index()
df[["name", "version", "purl"]].sort_values(by="name")
```

```python
# Cachito reported packages
df = cachi2_df[cachi2_df["purl"].str.contains("type=package", na=False)]
df = df[df["version"].notna()].reset_index()
df[["name", "version", "purl"]].sort_values(by="name")
```

```python
# Cachito reported std packages
df = cachi2_df[cachi2_df["purl"].str.contains("type=package", na=False)]
df = df[df["version"].isna()].reset_index()
df[["name", "version", "purl"]].sort_values(by="name")
```
